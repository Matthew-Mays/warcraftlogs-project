```python
import pandas as pd
```

## Acquire the data:

### We will be using pandas read_csv function to call our wowbgs.csv file into a dataframe.

```python
bgs_df = pd.read_csv('archive/wowbgs.csv')
bgs_df.head()
```

## Prepare the data:

### First there are some nulls to deal with so we will see just how many.

```python
bgs_df.info()
```

### The null values are in the win, lose, and bonus event columns. These are a categorical true/false that has been converted into 1/0, I will fill the null values with 0s.

```python
bgs_df = bgs_df.fillna(0)
```

```python
bgs_df.info()
```

### That's all the nulls dealt with. Next up, all the column names are capitalized and that can be annoying for calling them down the road. I will lower case them.

```python
bgs_df
```

```python
# Using list comprehension we can just use .lower on every column in df.columns
bgs_df.columns = [column.lower() for column in bgs_df.columns]
```

```python
bgs_df
```

### Much better, next up we need to deal with the capitalization in the observations, the battleground, code, faction, and class columns all have capitalized strings.

```python
# Testing first column trying to lowercase all values in battleground column
bgs_df['battleground'] = bgs_df['battleground'].str.lower()
bgs_df
```

```python
# Applying same solution to all other required columns
bgs_df['code'] = bgs_df['code'].str.lower()
bgs_df['faction'] = bgs_df['faction'].str.lower()
bgs_df['class'] = bgs_df['class'].str.lower()

bgs_df
```

### Now we wrap all of these into a function that will return the prepped dataframe.

```python
# This function will acquire and prepare the data for exploration
def get_wrangle_bgs_data():
    
    # Acquire the data
    bgs_df = pd.read_csv('archive/wowbgs.csv')
    
    # Fill categorically negative nulls with a value of 0
    bgs_df = bgs_df.fillna(0)
    
    # Setting all columns to lower case for convenience
    bgs_df.columns = [column.lower() for column in bgs_df.columns]

    # Lowercase all string rows for convenience
    bgs_df['battleground'] = bgs_df['battleground'].str.lower()
    bgs_df['code'] = bgs_df['code'].str.lower()
    bgs_df['faction'] = bgs_df['faction'].str.lower()
    bgs_df['class'] = bgs_df['class'].str.lower()
    
    return bgs_df
```

```python
new_bgs_df = get_wrangle_bgs_data()
new_bgs_df
```
